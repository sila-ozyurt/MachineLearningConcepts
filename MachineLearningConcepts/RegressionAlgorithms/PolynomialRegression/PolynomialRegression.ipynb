```python
import numpy as np
import pandas as pd
#in polynomial regression scaling is important
```

```python
data=pd.read_csv("insurance.csv")
df=data.copy()
df
```

```python
#no null values in it
df.isna().sum()
```

```python
df["region"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

le=LabelEncoder()
ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")


ohe_transform=ohe.fit_transform(df[["region"]])
ohe_transform.drop(columns=["region_northeast"],inplace=True)
ohe_transform

```

```python
df2=pd.concat([df,ohe_transform],axis=1).drop(columns=["region"])
df2
```

```python
df
```

```python
columns=["sex","smoker"]
df[columns]=df[columns].apply(lambda col:le.fit_transform(col))
df
```

```python
#instead of using le for binary values, we can simply encode it as below
#df2 was the version in which only onehotencoding applied

df2["sex"]=df2["sex"].replace({"female":0,"male":1})
df2["smoker"]=df2["smoker"].replace({"no":0,"yes":1})
df2.rename(columns={"sex":"isMale"},inplace=True)
df2
```

```python
df["sex"]=df["sex"].replace
```

```python
#lets split data as x and y
#predict charges
x=df2.drop(columns=["charges"])
x
```

```python
y=df2[["charges"]]
y
```

```python
#split data as train test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#standardize values
#in polynomial regression scaling is important
#we can use one of the scaler 

from sklearn.preprocessing import StandardScaler,MinMaxScaler

ss=StandardScaler()
mms=MinMaxScaler()

#we don't want the model to learn about test data so only transform test data not fit
x_train_standart_scaler=ss.fit_transform(x_train)
x_test_standart_scaler=ss.transform(x_test)

x_train_min_max_scaler=mms.fit_transform(x_train)
x_test_min_max_scaler=mms.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures

#degree 6
poly_reg=PolynomialFeatures(degree=2)

x_poly_train=poly_reg.fit_transform(x_train_standart_scaler)
x_poly_test=poly_reg.transform(x_test_standart_scaler)

lr.fit(x_poly_train,y_train)

y_predicted=lr.predict(x_poly_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score

mean_absolute_error(y_test,y_predicted)
```

```python
#it is overfitting
#decrease degree and try again
y_predicted_train=lr.predict(x_poly_train)
mean_absolute_error(y_train,y_predicted_train)
```

```python
r2_score(y_test,y_predicted)
```
