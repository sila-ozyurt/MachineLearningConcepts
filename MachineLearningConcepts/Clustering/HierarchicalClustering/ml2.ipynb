```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("musteriler.csv")
df=data.copy()
df
```

```python
#split data as x 
#turn it into numpy array 
#DO NOT FORGET 
x=df.iloc[:,3:].values
x
```

```python
from sklearn.cluster import AgglomerativeClustering
#affinity euclidean ,manhattan,precomputed
#linkage ward
#if linkage is ward affinity can only be euclidian

agg_cluster = AgglomerativeClustering(n_clusters=3,linkage="ward", metric="euclidean")

#lets show their classes
y_predicted=agg_cluster.fit_predict(x)
```

```python
#clusters according to 2 features 0 and 1
plt.scatter(x[y_predicted==0,0],x[y_predicted==0,1],s=100,color="red",label="cluster1")
plt.scatter(x[y_predicted==1,0],x[y_predicted==1,1],s=100,color="blue",label="cluster2")
plt.scatter(x[y_predicted==2,0],x[y_predicted==2,1],s=100,color="yellow",label="cluster3")
```

```python
#contrast it with k means algorithm

from sklearn.cluster import KMeans

results=list()

for i in range(1,11):
    k_means=KMeans(n_clusters=i,init="k-means++",random_state=123)
    k_means.fit(x)
    results.append(k_means.inertia_)
```

```python
#lets plot a dendogram to minimize inertia values

plt.plot(range(1,11),results)
```

```python
k_means=KMeans(n_clusters=4,init="k-means++",random_state=123)

y_predicted=k_means.fit_predict(x)

plt.scatter(x[y_predicted==0,0],x[y_predicted==0,1],s=10,color="red",label="cluster1")
plt.scatter(x[y_predicted==1,0],x[y_predicted==1,1],s=100,color="green",label="cluster2")
plt.scatter(x[y_predicted==2,0],x[y_predicted==2,1],s=100,color="purple",label="cluster3")
plt.scatter(x[y_predicted==3,0],x[y_predicted==3,1],s=100,color="yellow",label="cluster3")
```

```python
#DENDOGRAM

import scipy.cluster.hierarchy as sch

dendogram=sch.dendrogram(sch.linkage(x,method="ward"))
plt.show()
```
