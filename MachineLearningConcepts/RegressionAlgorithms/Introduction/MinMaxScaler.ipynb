```python
#in normalization, values distributes around [0,1]
#you can change the values by feature range
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("500hits.csv",encoding="latin1")
df=data.copy()
df.head()
```

```python
#drop player and cs column which is not needed
df=df.drop(columns=["CS","PLAYER"],axis=1)
df
```

```python
df.info()
```

```python
df.describe().round(3)
```

```python
x=df.iloc[:,0:13]
x
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler

min_max_scaler=MinMaxScaler(feature_range=(0,1))

x=min_max_scaler.fit_transform(x)
x
```

```python
#x is an array
type(x)
df.columns
```

```python
#switch x to a df

x=pd.DataFrame(x,columns=['YRS', 'G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'BB', 'SO', 'SB',
       'BA'])
```
