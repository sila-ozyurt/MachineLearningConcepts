```python
import numpy as np
import pandas as pd
```

```python
data= {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
      'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
      'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small']}
```

```python
df=pd.DataFrame(data=data)
df.head()
```

```python
#lets check out different values of city columns
df["city"].unique()
```

```python
#number of different values are not much, so we can use one hot encoding
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")
encoded_data=ohe.fit_transform(df[["city"]])
encoded_data
```

```python
#drop dummy variables
encoded_data.drop(columns=["city_Tampa"],inplace=True)
df.drop(columns=["city"],inplace=True)
```

```python
#real df still didn't change
df.head()
```

```python
#concat 2 df
df=pd.concat([df,encoded_data],axis=1)
df
```
