```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("veriler.csv")
df=data.copy()
df
```

```python
#check out unique vals
df["ulke"].unique()
```

```python
#one hot encode with ulke column

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")
encoded_data=ohe.fit_transform(df[["ulke"]])
encoded_data
```

```python
#drop dummy variables

encoded_data.drop(columns=["ulke_us"],inplace=True)
df.drop(columns=["ulke"],inplace=True)
```

```python
#concat encoded data and df

df=pd.concat([df,encoded_data],axis=1)
df
```

```python
# cinsiyet column does not affect boy kilo or yas
df
```

```python
#split data as x and y
x=df.drop("cinsiyet",axis=1)
x
```

```python
y=df["cinsiyet"]
y
```

```python
#trains and test are row-oriented splits
#x and y are splits for dependent and independent variable
#x independent var
#y dependent var

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
#MinMaxScaler is typically preferred to preserve the data distribution,original distribution is crucial
# StandardScaler is used in algorithms where differences and variances among features are significant,less outlier sensitivity

from sklearn.preprocessing import MinMaxScaler

mms=MinMaxScaler()
scaled_x_train=mms.fit_transform(x_train)
scaled_x_test=mms.fit_transform(x_test)
```
