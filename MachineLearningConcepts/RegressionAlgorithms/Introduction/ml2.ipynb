```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#take the data from csv
data=pd.read_csv("eksikveriler.csv")
df=data.copy()
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#fill in the nan values
age=df.iloc[:,1:4].values
age
```

```python
#fill in the nan values with mean strategy
from sklearn.impute import SimpleImputer

simple_imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
imputed_data=simple_imputer.fit_transform(age)
imputed_data
```

```python
#encode country values
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")
encoded_data=ohe.fit_transform(df[["ulke"]])
encoded_data
```

```python
#it is still not updated
df.head()
```

```python
#gather 2 df
df=pd.concat([df,encoded_data],axis=1)
df
```

```python
#get rid of dummy var
df.drop(columns=["ulke"],inplace=True)
df
```
