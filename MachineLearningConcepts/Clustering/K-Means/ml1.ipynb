```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("musteriler.csv")
df=data.copy()
df
```

```python
#split data as x 
x=df.iloc[:,3:]
x
```

```python
from sklearn.cluster import KMeans

k_means=KMeans(n_clusters=3,init="k-means++")
k_means.fit(x)
```

```python
k_means.cluster_centers_
```

```python
results=list()

#lets calculate wcss values
for i in range(1,10):
    k_means=KMeans(n_clusters=i,init="k-means++",random_state=123)
    k_means.fit(x)
    results.append(k_means.inertia_)
```

```python
plt.plot(range(1,10),results)
```
