```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("maaslar.csv")
df=data.copy()
df
```

```python
#visualize
import seaborn as sbn

sbn.pairplot(data=df)
```

```python
plt.scatter(x="Egitim Seviyesi",y="maas",data=df)
plt.show()
```

```python
#split data as x y
x=df[["Egitim Seviyesi"]]
x
```

```python
y=df[["maas"]]
y
```

```python
#split data as train and test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
#scale the data
#outliers are not so crucial here so use standard scaling

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()

x_train_scaled=ss.fit_transform(x_train)
#we should not teach test data to the system
x_test_scaled=ss.transform(x_test)
```

```python
#use support vector machine 

from sklearn.svm import SVR
#linear poly rbf sigmoid

support_vector_reg=SVR(kernel='poly')
support_vector_reg.fit(x_train_scaled,y_train)
y_predicted=support_vector_reg.predict(x_test_scaled)
y_predicted
```
