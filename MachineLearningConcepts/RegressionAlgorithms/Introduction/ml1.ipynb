```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create df from the csv values
data=pd.read_csv("eksikveriler.csv")
df=data.copy()
df
```

```python
#check out for the null values
df.isnull().sum()
```

```python
#to fill nan values, placing in boy,kilo,yas columns
#adding values it turns df to a num arr
age=df.iloc[:,1:4].values
age
```

```python
#we can do both seperately(fit() transform() )
# from sklearn.impute import SimpleImputer
# #mean median most_frequent

# imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
# imputer=imputer.fit(age[:,1:4])
# age[:,1:4]=imputer.transform(age[:,1:4])
# age


#and together (fit_transform() )
from sklearn.impute import SimpleImputer

s_imp=SimpleImputer(missing_values=np.nan,strategy="median")
imputed_data=s_imp.fit_transform(age)
imputed_data
```

```python
#its type is num arr
type(imputed_data)
```

```python
#switch it to a df

imputed_data=pd.DataFrame()
imputed_df=pd.DataFrame(imputed_data,columns=["boy","kilo","yas"])
imputed_df
```

```python
#merge it to the real df
df=df.fillna(imputed_df)
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")
ulke_encoded=ohe.fit_transform(df[["ulke"]])
ulke_encoded
```

```python
#concat with encoded column
df=pd.concat([df,ulke_encoded],axis=1).drop("ulke",axis=1)
```
