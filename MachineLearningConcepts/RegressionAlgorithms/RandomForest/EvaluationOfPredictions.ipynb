```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("maaslar.csv")
df=data.copy()
df
```

```python
#split data as x y
x=df[["Egitim Seviyesi"]]
x
```

```python
y=df[["maas"]]
y
```

```python
#lets try random forest regressor algorithm first
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor(random_state=0,n_estimators=10)
rfr.fit(x,y)
```

```python
from sklearn.metrics import r2_score

r2_score(y,rfr.predict(x))
```

```python
k=x-0.4
z=x+0.5
```

```python
r2_score(y,rfr.predict(z))
```

```python
r2_score(y,rfr.predict(k))
```

```python
#lets try decision tree 
#dt always returns the same interval for all values
#so it s normal 1 to be given in r2_score 

from sklearn.tree import DecisionTreeRegressor

dt_r=DecisionTreeRegressor(random_state=0)
dt_r.fit(x,y)
```

```python
r2_score(y,dt_r.predict(x))
```

```python
#lets try support vector regression (svr)
#it is must to use scaler in SVR

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()

x_scaled=ss.fit_transform(x)
y_scaled=ss.fit_transform(y)
```

```python
#SVR
from sklearn.svm import SVR
#kernel= linear,poly,rbf(radial basis function),sigmoid
svr=SVR(kernel="rbf")
svr.fit(x_scaled,y_scaled)
```

```python
r2_score(y_scaled,svr.predict(x_scaled))
```

```python
#lets check out polynomial regression

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)

lr2=LinearRegression()

lr2.fit(x_poly,y)
```

```python
r2_score(y,lr2.predict(x_poly))
```

```python
#lets go on with linear regression
lr=LinearRegression()
lr.fit(x,y)
```

```python
r2_score(y,lr.predict(x))
```

```python
print("linear= ",r2_score(y,lr.predict(x)))
print("polynomial= ",r2_score(y,lr2.predict(x_poly)))
print("SVR radial basis func= ",r2_score(y_scaled,svr.predict(x_scaled)))
print("decision tree= ",r2_score(y,dt_r.predict(x)))
print("random forest tree= ",r2_score(y,rfr.predict(x)))
```
