```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("veriler.csv")
df=data.copy()
print(df.shape)
```

```python
#lets encode cinsiyet column
#one hot encoder

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")
ohe_transform=ohe.fit_transform(df[["cinsiyet"]])
ohe_transform
```

```python
#lets delete dummy columms and concat 2 df

df=pd.concat([df,ohe_transform],axis=1).drop(columns=["cinsiyet","cinsiyet_e"])
df
```

```python
#lets encode ulke column

ohe_transform=ohe.fit_transform(df[["ulke"]])
ohe_transform
```

```python
#lets concat 

df=pd.concat([df,ohe_transform],axis=1).drop(columns=["ulke"])
df
```

```python
y=df["boy"]
df.drop(columns="boy",inplace=True)
```

```python
#check out which variable is corrupting model, measuring their p values(max p val)
import statsmodels.api as sm

#try to add b0 values in here [   Y=B0+B1X1+B2X2+B3X3+ERROR_RATE ]
X=np.append(arr=np.ones((22,1)).astype(int),values=df,axis=1)
X
```

```python
#create a list for each column in df and eliminate them to reach most proper columns for regression

X_list=df.iloc[:,[0,1,2,3,4,5]].values
X_list
```

```python
type(X_list)
```

```python
X_list=np.array(X_list,dtype=float)
#measure the impact of each column over 'boy' column
model=sm.OLS(y,X_list).fit()
model.summary()
```
