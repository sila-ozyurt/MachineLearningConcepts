```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("veriler.csv")
df=data.copy()
df.head()
```

```python
#encode cinsiyet column
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")
ohe_transform=ohe.fit_transform(df[["cinsiyet"]])
ohe_transform
```

```python
#delete one column either cinsiyet_e or cinsiyet_k. These are dummy variables
df=pd.concat([df,ohe_transform],axis=1).drop(columns=["cinsiyet_e","cinsiyet"])
df
```

```python
df
```

```python
#encode ulke column
ohe_transform=ohe.fit_transform(df[["ulke"]])
ohe_transform
```

```python
#delete one of the columns to prevent dummy variable track

df=pd.concat([df,ohe_transform],axis=1).drop(columns=["ulke_us"])
```

```python
#drop ulke column
df.drop(columns=["ulke"],inplace=True)
df
```

```python
#split data for x
other_val=df.iloc[:,0:3]
other_val=pd.concat([other_val,df.iloc[:,4:6]],axis=1)
other_val
```

```python
#split data for y
y_val=df.iloc[:,3]
y_val
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(other_val,y_val,test_size=0.33,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(x_train,y_train)
y_predicted=lr.predict(x_test)


```
